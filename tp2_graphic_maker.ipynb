### Nesse arquivo todos os dados gerados e salvos são transformados em gráficos.

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("christofides.csv")
df = pd.DataFrame(data)
df = df.iloc[:-8]

name = "Twice Around The Tree"

# Ordenar o DataFrame pelo tempo de execução
df = df.sort_values(by='Tempo de execução')

# Ajustar o tamanho da figura e espaçamento vertical
fig, ax = plt.subplots(figsize=(16, 4))

# Criar o gráfico de barras com eixo y logarítmico
bars = ax.bar(df['Nome do Arquivo'], df['Tempo de execução'], color='blue')
ax.set_yscale('log')  # Adicionar escala logarítmica ao eixo y

# Adicionar os valores no topo de cada barrinha de maneira inclinada
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval * 1.25, f"{round(yval, 1)} s", ha='center', va='bottom', rotation=90)

# Rotacionar os rótulos em 45 graus para melhor legibilidade
ax.set_xticks(df['Nome do Arquivo'])
ax.set_xticklabels(df['Nome do Arquivo'], rotation=45, ha="right")

# Remover a borda superior
ax.spines['top'].set_visible(False)

ax.set_xlabel('Dataset')
ax.set_ylabel('Tempo de Execução (Escala Logarítmica)')
#ax.set_title('Tempo de Execução para Cada Dataset\n'+name )
fig.tight_layout()  # Ajusta automaticamente o layout para evitar cortes
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Calcular o número de vértices para cada entrada
df['Num_Vertices'] = df['Vertices'].apply(lambda x: len(eval(x)) + 1)

# Ajustar o tamanho da figura e espaçamento vertical
fig, ax = plt.subplots(figsize=(7, 4))

# Criar o gráfico de dispersão
scatter = ax.scatter(df['Num_Vertices'], df['Tempo de execução'], color='blue', alpha=0.7, label='Pontos de Dados')

# Ajustar uma curva polinomial de grau 2
coefficients = np.polyfit(df['Num_Vertices'], df['Tempo de execução'], 2)
polyfit_line = np.poly1d(coefficients)
x_fit = np.linspace(df['Num_Vertices'].min(), df['Num_Vertices'].max(), 100)
y_fit = polyfit_line(x_fit)

# Adicionar a linha de ajuste polinomial com a legenda
label_polyfit = f'Ajuste Polinomial'
ax.plot(x_fit, y_fit, color='red', label=label_polyfit)

# Adicionar rótulos e título
ax.set_xlabel('Número de Vértices')
ax.set_ylabel('Tempo de Execução')
#ax.set_title('Número de Vértices por Tempo de Execução\n'+name)

# Adicionar grade para melhor visualização
#ax.grid(True, linestyle='--', alpha=0.6)

# Adicionar legenda
ax.legend()

fig.tight_layout()  # Ajusta automaticamente o layout para evitar cortes

# Imprimir coeficiente de correlação
correlation_coefficient = np.corrcoef(df['Num_Vertices'], df['Tempo de execução'])[0, 1]
print(f'Coeficiente de Correlação (r): {correlation_coefficient}')

plt.show()
```
